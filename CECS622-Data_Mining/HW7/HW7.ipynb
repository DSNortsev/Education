````python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

x = ['1996', '1997', '1998', '1999', '2001', '2001']
y = [7, 5, 7, 9, 5, 7]

x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, y, color='green')
plt.xlabel("Year")
plt.ylabel("A")

plt.xticks(x_pos, x)
print('(a) Show a bar chart for the variable A') 
plt.show() 
````
