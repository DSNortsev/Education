```python
import pandas as pd

df = pd.read_excel('location.xlsx')
df.head()  
```

```python
from bokeh.io import show, output_file
from bokeh.plotting import figure

output_file("states_pro_security.html")

states = list(df[df['Topic'] == 'pro-security']['Location'])
counts = list(df[df['Topic'] == 'pro-security']['Counts'])

sorted_states = sorted(states, key=lambda x: counts[states.index(x)])

p = figure(
  y_range=sorted_states,
  title = 'Public Safety advocates by states',
  x_axis_label ='Counts',
  plot_width=800,
  plot_height=600,
  tools=""
)

p.hbar(
    y=states,
    right=counts,
    left=0,
    height=0.4,
    color='green',
    fill_alpha=0.5
)

show(p)
```
